```python
import os
from django.db.models import Max
import pandas as pd
#sys.path.append('/home/eddyod/programming/activebrainatlas')
from brain.models import RawSection
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
```

```python
sections = RawSection.objects.filter(prep_id='DK43').order_by('-slide_physical_id', '-scene_number')
```

```python
#sections = RawSection.objects.filter(prep_id='DK43').order_by('-slide_physical_id', '-scene_number','channel')[:12]
raw_sections = RawSection.objects.filter(prep_id='DK43').order_by('-section_number', 'channel')[:12000]
channels = raw_sections.aggregate(Max('channel'))
channels = channels['channel__max']

new_section_number = 0
for i, section in enumerate(raw_sections):
    if i % channels == 0:
        new_section_number += 1
    if new_section_number > 0:
        print(section.source_file, new_section_number, section.channel)
    section.section_number = new_section_number
    section.save()
```
